```python
import pandas as pd
```

```python
df = pd.read_excel('edges.xlsx')
df
```

```python
df = df.drop('Unnamed: 0', axis=1)
df
```

```python
count = df.iloc[0]['distance']
mincount = float('inf')
index = []
for i in range(1, len(df)):
    if df.iloc[i]['x1'] == df.iloc[i-1]['x1']:
        count += df.iloc[i]['distance']
    else:
        if mincount >= count:
            mincount = count
            index.append((mincount, df.iloc[i-1]['x1']))
            count = df.iloc[0]['distance']
print(index)
```
