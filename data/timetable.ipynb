```python
import pandas as pd
from datetime import datetime
```

```python
sheet_1 = pd.read_excel('расписание_и_причалы.xlsx', sheet_name='Причалы')
sheet_2 = pd.read_excel('расписание_и_причалы.xlsx', sheet_name='Расписание0307')
timetable = pd.merge(sheet_1, sheet_2, on='ID причала')
timetable
```

```python
timetable_rename = timetable.drop('Название причала_y', axis=1)
new_columns = list(sheet_1.iloc[0]) + list(sheet_2.iloc[0])
del new_columns[7]
del new_columns[7]
timetable_rename.columns = new_columns
timetable_rename
```

```python
timetable_rename['Начало расписания'] = pd.to_datetime(timetable_rename['Начало расписания'])
timetable_rename['Окончание расписания'] = pd.to_datetime(timetable_rename['Окончание расписания'])
timetable_update = timetable_rename[timetable_rename["changed_schedule.type"] != 'Отмена']
```

```python
timetable_today = timetable_update[(timetable_update['Окончание расписания'] >= pd.to_datetime('07.03.2024')) & (timetable_update['Начало расписания'] <= pd.to_datetime('07.03.2024'))]
```

```python
def plus_two_minutes(time_list):
    return [(datetime.strptime(f'{t}', "%H:%M:%S") + pd.Timedelta(minutes=2)).time() for t in time_list if pd.notna(t)]
```

```python
grouped = timetable_today.groupby(['Docs.name', 'position.berth_letter'])
start = grouped['Причаливание'].apply(list)
end = grouped['Отход'].apply(list)

past_start = start.apply(plus_two_minutes)
past_end = end.apply(plus_two_minutes)
```

```python
data = {
    'Начало подхода': start,
    'Конец подхода': past_start,
    'Начало отхода': end,
    'Конец отхода': past_end
}
df = pd.DataFrame(data)
df
```

```python
df.to_excel('timetable.xlsx', index=True)
```
